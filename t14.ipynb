```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st

x = np.array([-1.11, -6.10, 2.42])
y = np.array([-2.29, -2.91])
n = len(x)
m = len(y)

var_x = 2
var_y = 1
alpha = 0.05

mean_x = np.mean(x)
mean_y = np.mean(y)
print(f"Среднее x: {mean_x:.3f}")
print(f"Среднее y: {mean_y:.3f}")

d_stat = (mean_x - mean_y) / np.sqrt(var_x / n + var_y / m)
critical = st.norm.ppf(1 - alpha)
print(f"Статистика критерия (Δ): {d_stat:.3f}")
print(f"Критическая область: Δ > {critical:.3f}")
```

```python
theta_vals = np.linspace(-1, 5, 1000)
quantile = st.norm(0, np.sqrt(var_x / n + var_y / m)).ppf(1 - alpha)
power = 1 - st.norm(theta_vals, np.sqrt(var_x / n + var_y / m)).cdf(quantile)

plt.figure(figsize=(10, 5))
plt.plot(theta_vals, power, color="blue")
plt.title("График мощности")
plt.xlabel("Параметр сдвига (μ_x - μ_y)")
plt.ylabel("Мощность")
plt.grid(alpha=0.3)
plt.show()
```
