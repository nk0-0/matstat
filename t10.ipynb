```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import math

np.random.seed(11)
n = 100
# α χ2 Δ
```

```python
a = 0.05
n = 100
numbers = np.ones(10)
arr = np.array([5, 8, 6, 12, 14, 18, 11, 6, 13, 7])
```

# а)

```python
F_emp = np.array([arr[:i].sum() for i in range(len(arr) + 1)]) / n
F_t = np.array([numbers[:i].sum() for i in range(len(numbers))]) / 10

delta = float(np.sqrt(n) * max(max(abs(F_emp[i] - F_t[i]), abs(F_emp[i + 1] - F_t[i])) for i in range(10)))

print(f"Δ = {delta:.3f}")
```

```python
def kolmogorov_sum(x):
    res = 1.0
    for i in range(1, 1001):
        res += 2 * (-1) ** i * np.exp(-2 * i*i * x*x)
    return res
```

```python
p_value = 1 - kolmogorov_sum(delta)
print(f"p-value = {p_value:.3f}")
```

```python
print(f"p-value < α = 0.05 => Отвергаем H0")
```

# б)

```python
def normal_distr(x, mean, sigma):
    return 0.5 * (1 + math.erf((x - mean)/(np.sqrt(2) * sigma)))

def get_delta_arr(mean, sigma, N, n):
    delta_arr  = list()
    for i in range(N):
        sample = np.random.normal(mean, sigma, n)
        sorted_sample = sorted(sample)

        bootstrap_F_emp = [i / n for i in range(n + 1)]
        bootstrap_mean = np.mean(sample)
        bootstrap_sigma = np.std(sample) * n /(n - 1)

        delta_arr.append(np.max([np.sqrt(n) * max(np.abs(normal_distr(sorted_sample[j], bootstrap_mean, bootstrap_sigma) - bootstrap_F_emp[j]), \
                                     np.abs(normal_distr(sorted_sample[j], bootstrap_mean, bootstrap_sigma) - bootstrap_F_emp[j+1])) \
                                        for j in range(len(sorted_sample))]))
        
    delta_arr = np.array(delta_arr)
    return delta_arr


N = 10000
n = 100

mean = 0
for i in range(len(arr)):
    mean += (arr[i]/n)*i

disp = 0

for i in range(len(arr)):
    disp += arr[i]*(i - mean)**2 / (n - 1)

sigma = np.sqrt(disp)

x = [i for i in range(10)]

delta_ = np.max([np.sqrt(n) * max(np.abs(normal_distr(x[i], mean, sigma) - F_emp[i]),np.abs(normal_distr(x[i], mean, sigma) - F_emp[i+1])) for i in range(10)])

delta_arr = get_delta_arr(mean, sigma, N, n)
p_val = len(delta_arr[delta_arr >= delta_]) / N
print(f"p-value = {p_val}")
```

```python
print(f"p-value < α = 0.05 => Отвергаем H0")
```
